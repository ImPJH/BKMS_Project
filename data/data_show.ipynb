# crime_preprocessed.csv
        - CMPLNT_NUM (사건번호)
        - ADDR_PCT_CD (precinct)
        - BORO_NM (자치구 이름) : Bronx, Queens, Brooklyn, Manhattan, Staten Island, (null)

        - Latitude (위도)
        - Longitude (경도)
        - Lat_Lon (위도,경도 pair)
